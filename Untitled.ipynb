```python
from transformers import pipeline
import pandas as pd
from urllib.parse import urlparse
import requests
from PIL import Image
from io import BytesIO
```

```python
class NewsModel:

    def __init__(self):
        # Инициализация модели для анализа настроений
        self.sentiment_model = pipeline("sentiment-analysis")

    def preprocess_text(self, text):
        """
        Предобработка текста новостей:
        - Убираем ссылки, специальные символы и приводим текст к нижнему регистру.
        """
        text = re.sub(r'http\S+', '', text)  # Удаляем ссылки
        text = re.sub(r'[^a-zA-Zа-яА-Я\s]', '', text)  # Удаляем специальные символы
        text = text.lower().strip()  # Приводим текст к нижнему регистру
        return text

    def analyze_sentiment(self, text):
        """
        Анализ тональности новости:
        Возвращает числовую оценку настроения, где:
        - Положительные новости -> положительное значение.
        - Отрицательные новости -> отрицательное значение.
        """
        preprocessed_text = self.preprocess_text(text)
        sentiment = self.sentiment_model(preprocessed_text)[0]  # Получаем результат анализа
        score = sentiment['score'] if sentiment['label'] == 'POSITIVE' else -sentiment['score']
        return score

    def process_news(self, news_list):
        """
        Обрабатывает список новостей и возвращает DataFrame с текстами новостей и их оценками настроений.
        """
        news_data = []
        for news in news_list:
            score = self.analyze_sentiment(news)
            news_data.append({'news': news, 'sentiment_score': score})

        return pd.DataFrame(news_data)

    
    def get_sentiment_scores(self, news_data):
        sentiments = self.model(news_data)
        scores = [sentiment['score'] if sentiment['label'] == 'POSITIVE' else -sentiment['score'] for sentiment in sentiments]
        return scores
```
